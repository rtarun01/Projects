## Returns measures

```python
# simple return 
df1['simple return']=df1['Adj Close'].pct_change()
df1
```

```python
# log return 
df1['log return']=np.log(df1['Adj Close']/df1['Adj Close'].shift(1))
df1
```

```python
# cumulative return
df1['simple return_prod']=df1['simple return'].add(1).cumprod()-1
df1['log return_sum_exp']=np.exp(df1['log return'].cumsum())-1
df1
```

```python
# total return 
df1.loc['2018-12-31','Adj Close']/df1.loc['2010-01-04','Adj Close']-1
```

```python
# total return 
df1.iloc[-1,4]/df1.iloc[0,4]-1
```

```python

df1.drop(labels=['simple return','log return','log return_sum_exp','simple return_prod'], axis='columns', inplace=True)
df1
```

## Rolling statisitcs 

```python
# log return
df1['return'] = np.log(df1['Adj Close'] / df1['Adj Close'].shift(1))
df1
```

```python
# rolling minimum value 
df1['min']=df1['Adj Close'].rolling(window=5).min()
df1.tail()
```

```python
# rolling mean value 
df1['mean']=df1['Adj Close'].rolling(5).mean()
df1.tail()
```

```python
# rolling standard deviation
df1['std']=df1['Adj Close'].rolling(5).std()
df1.tail()
```

```python
# rolling median value
df1['median']=df1['Adj Close'].rolling(5).median()
df1.tail()
```

```python
# rolling maximum value 
df1['max']=df1['Adj Close'].rolling(5).max()
df1.tail()
```

```python
# delete columns 
df1.drop(labels=['min','mean','std','median','max'],axis='columns',inplace=True)
df1
```

## Moving averages 

```python
# one week
df1['5-day'] = df1['Adj Close'].rolling(5).mean()
df1
```

```python
# one month 
df1['21-day'] = df1['Adj Close'].rolling(21).mean()
df1
```

```python
# 2 month
df1['42-day'] = df1['Adj Close'].rolling(42).mean()
df1
```

```python
# 1 year
df1['252-day'] = df1['Adj Close'].rolling(252).mean()
df1
```

```python
# drop NaN
df1.dropna(inplace=True)
df1
```

```python
# delete columns 
df1.drop(labels=['High','Low','Open','Close','Volume'],axis='columns',inplace=True)
df1
```

```python
df1[['Adj Close','5-day','21-day']].plot(figsize=(20, 10))
plt.show()
```

```python
df1[['Adj Close','42-day','252-day']].plot(figsize=(15, 10))
plt.show()
```

## Trading strategy 1

```python
df1['position1'] = np.where(df1['5-day'] > df1['21-day'], 1 , -1)
df1
```

```python
# daily returns for strategy 1
df1['strategy1'] = df1['position1'].shift(1) * df1['return']
df1
```

```python
# cumulative returns for the benchmark 
df1['return_cum']=np.exp(df1['return'].cumsum())-1
df1
```

```python
# same  
df1['return'].cumsum().apply(np.exp)-1
```

```python
# cumulative returns for strategy 1
df1['strategy1_cum']=np.exp(df1['strategy1'].cumsum())-1
df1
```

```python
# or
df1['strategy1'].cumsum().apply(np.exp)-1
```

```python
# cumulative returns for the benchmark and strategy 1
df1[['return_cum', 'strategy1_cum']].plot(figsize=(20, 10))
plt.show()
```

```python
# or
df1[['return','strategy1']].cumsum().apply(np.exp).sub(1).plot(figsize=(20, 10))
plt.show()
```

## Defining a function

#### Download data from Yahoo Finance

```python
def strategy(ticker,n1,n2):
    df = yf.download(ticker, start="2010-01-01",end="2019-01-01")[['Adj Close']]
    df['return'] = np.log(df['Adj Close'] / df['Adj Close'].shift(1))
    df['mv1'] = df['Adj Close'].rolling(n1).mean()
    df['mv2'] = df['Adj Close'].rolling(n2).mean()
    df.dropna(inplace=True)
    df['position'] = np.where(df['mv1'] > df['mv2'], 1 , -1)
    df['strategy'] = df['position'].shift(1) * df['return']
    df[['return','strategy']].cumsum().apply(np.exp).sub(1).plot(figsize=(20, 10))
    plt.show()
```

```python
strategy('AAPL',42,252)
```
